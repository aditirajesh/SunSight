```python
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt 
import numpy as np
```

Importing csv file

```python
dataset = pd.read_csv("/Users/aditirajesh/Desktop/program_files/python/python_6sem/DATA_PLANTS/plant1/data1_1.csv")
df = pd.DataFrame(columns=['DATE','AMBIENT_TEMP','MODULE_TEMP','SUN_HOURS','YIELD'])

dataset[['DATE', 'TIME']] = dataset.DATE_TIME.str.split(" ", expand = True)
dataset.drop(["DATE_TIME","DC_POWER","AC_POWER"],axis=1,inplace=True)
dataset = dataset[['DATE','TIME','AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION','DAILY_YIELD']]
print(dataset)

print(dataset.columns)
```

```python

for date in dataset['DATE'].unique():
    ambient_temperatures = list(dataset.loc[dataset['DATE'] == date, 'AMBIENT_TEMPERATURE'])
    module_temperatures = dataset.loc[dataset['DATE'] == date, 'MODULE_TEMPERATURE']
    irradiations = dataset.loc[dataset['DATE'] == date, 'IRRADIATION']

    ambient_temperatures = sum(ambient_temperatures)/len(ambient_temperatures)
    module_temperatures = sum(module_temperatures)/len(module_temperatures)
    irradiations = sum(irradiations)/4

    daily_yield = dataset.loc[(dataset['DATE'] == date) & (dataset['TIME'] == '18:30:00'),'DAILY_YIELD']

    df = df._append({'DATE':date,
                    'AMBIENT_TEMP':ambient_temperatures,
                    'MODULE_TEMP':module_temperatures,
                    'SUN_HOURS':irradiations,
                    'YIELD':daily_yield},ignore_index=True)

print(df)

```
